```python
import os
import pandas as pd
import numpy as np
```

```python
PATH = '/content/drive/MyDrive/23-2_DSL/EDA' # Dataset Directory Path 지정
```

```python
names = ['계량경제학(1)','계약및조직이론'] # Dataset Directory 내 강의명 제시
```

### Add Column Index

```python
for name in names:
    for file in os.listdir(os.path.join(PATH, name)):
        file_csv = file.split('.')[0] + '.csv'
        if file.startswith('202'):
            with open(os.path.join(PATH, name, file), "r") as file_name:
                first_line = file_name.readline()
                if first_line[0] == '1' or first_line[0] == '0':
                    DF = pd.read_csv(os.path.join(PATH, name, file), delimiter = '\t', header = None)
                    DF.columns = ["순위", "마일리지", "특수교육대상자(장애학생)여부", "전공자/복수전공자(전공자정원포함여부)", "신청과목수", "졸업신청", "초수강여부", "총이수학점비율", "직전학기이수학점비율", "학년", "수강여부", "비고"]
                    DF.to_csv(os.path.join(PATH, name, file_csv), index = False)


        os.remove(os.path.join(PATH, name, file))
        print(file, 'is switched to csv format!')
```

```python
for name in names:
    for file in os.listdir(os.path.join(PATH, name)):
        if file.startswith('202'):
            DF = pd.read_csv(os.path.join(PATH, name, file), delimiter = ',')

            if DF['특수교육대상자(장애학생)여부'][0] == 'Y' or DF['특수교육대상자(장애학생)여부'][0] == 'N':
                DF['특수교육대상자(장애학생)여부'] = DF['특수교육대상자(장애학생)여부'].fillna('N').map({'N' : 0, 'Y' : 1}).astype(int)

            if DF['전공자/복수전공자(전공자정원포함여부)'][0] != 1 and DF['전공자/복수전공자(전공자정원포함여부)'][0] != 2 and DF['전공자/복수전공자(전공자정원포함여부)'][0] != 3 and DF['전공자/복수전공자(전공자정원포함여부)'][0] != 4:
                DF['전공자/복수전공자(전공자정원포함여부)'] = DF['전공자/복수전공자(전공자정원포함여부)'].fillna('N(N)').map({'N(N)' : 1, 'N(Y)' : 2, 'Y(N)' : 3, 'Y(Y)' : 4}).astype(int)

            if DF['졸업신청'][0] != 0 and DF['졸업신청'][0] != 1:
                DF['졸업신청'] = DF['졸업신청'].fillna('N').map({'N' : 0, 'Y' : 1}).astype(int)

            if DF['초수강여부'][0] != 0 and DF['초수강여부'][0] != 1:
                DF['초수강여부'] = DF['초수강여부'].fillna('N').map({'N' : 0, 'Y' : 1}).astype(int)

            if DF['수강여부'][0] != 0 and DF['수강여부'][0] != 1:
                DF['수강여부'] = DF['수강여부'].fillna('N').map({'N' : 0, 'Y' : 1}).astype(int)
            DF.to_csv(os.path.join(PATH, name, file), index = False)

            print(file, 'is updated !')
```
