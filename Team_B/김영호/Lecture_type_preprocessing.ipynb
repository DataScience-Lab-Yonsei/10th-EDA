```python
# 마일리지 커트 확인

def mileage_cut(data):
    # data: pandas.DataFrame
    # return: int, 난이도(수강신청 실패 최대 마일리지)
    # --------------------

    data = data[data.iloc[:, -1] == 0]      # 수강신청 실패
    if len(data) == 0:
        # 수강신청 실패가 없음
        # 1 마일리지만 넣어도 수강신청 성공 -> 수강신청 실패 최대 마일리지 = 1
        max_mileage = 1
    else:
        # 수강신청 실패 최대 마일리지
        max_mileage = max(data['마일리지'])

    return int(max_mileage)

# 전공정원 포함 마일리지 커트
def mileage_cut_m(data):
  m_data = data[data.iloc[:, 2] == 4]
  if len(m_data) == 0:
    return(-1) #만약 비전공자만 신청했을 경우 결측처리를 위해 -1로 return
  else:
    m_data_fail = m_data[data.iloc[:, -1]==0]

    if len(m_data_fail) == 0:
      max_mileage = 1
    else:
      max_mileage = max(m_data_fail['마일리지'])

    return int(max_mileage)

# 전공정원 미포함 마일리지 커트
def mileage_cut_um(data):
  um_data = data[data.iloc[:, 2] != 4]
  if len(um_data) == 0:
    return(-1) #만약 전공자만 신청했을 경우 결측처리를 위해 -1로 return
  else:
    um_data_fail = um_data[data.iloc[:, -1]==0]

    if len(um_data_fail) == 0:
      max_mileage = 1
    else:
      max_mileage = max(um_data_fail['마일리지'])

    return int(max_mileage)
```

```python
# 수강성공 / 수강신청자 수
def prob_success(data):
  s_data = data[data.iloc[:, -1] == 1]
  f_data = data[data.iloc[:, -1] == 0]
  prob = len(s_data)/(len(s_data) + len(f_data))

  return(round(prob, 3))

# 전공자 수강성공 / 수강신청자 수
def prob_success_m(data):
  data = data[data.iloc[:, 2] == 4]
  if len(data) == 0:
    return(-1)
  else:
    s_data = data[data.iloc[:, -1] == 1]
    f_data = data[data.iloc[:, -1] == 0]
    prob = len(s_data)/(len(s_data) + len(f_data))

    return(round(prob, 3))


# 비전공자 수강성공 / 수강신청자 수
def prob_success_um(data):
  data = data[data.iloc[:, 2] != 4]
  if len(data) == 0:
    return(-1)
  else:
    s_data = data[data.iloc[:, -1] == 1]
    f_data = data[data.iloc[:, -1] == 0]
    prob = len(s_data)/(len(s_data) + len(f_data))

    return(round(prob, 3))
```

```python
final = pd.DataFrame()
year_list = []
semester_list = []
lecture_type_list = []
prof_list = []
name_list = []
class_list = []
mileage_cut_list = []
mileage_cut_m_list = []
mileage_cut_um_list = []
prob_success_list = []
prob_success_m_list = []
prob_success_um_list = []

for name in folders:
    for file in os.listdir(os.path.join(PATH, name)):
        filename_list = file.split('_')
        print(file, "starts")
        year_list.append(int(filename_list[0]))
        semester_list.append(int(filename_list[1]))
        lecture_type_list.append(str(filename_list[2]))
        prof_list.append(str(filename_list[3]))
        name_list.append(str(filename_list[4].replace(".csv", "")))
        if(len(filename_list)==6):
          class_list.append(str(filename_list[5].replace(".csv", "")))
        else:
          class_list.append("None")
        df = pd.read_csv(os.path.join(PATH, name, file), delimiter = ',')
        if(file == "2023_1_대면_김숙경_선형대수.csv"):
          print(df)
        #df['수강여부'] = df['수강여부'].astype(int)
        #df['전공자/복수전공자(전공자정원포함여부)']= df['전공자/복수전공자(전공자정원포함여부)'].astype(int)
        #df['마일리지'] = df['마일리지'].astype(int)
        mileage_cut_list.append(int(mileage_cut(df)))
        mileage_cut_m_list.append(int(mileage_cut_m(df)))
        mileage_cut_um_list.append(int(mileage_cut_um(df)))
        prob_success_list.append(float(prob_success(df)))
        prob_success_m_list.append(float(prob_success_m(df)))
        prob_success_um_list.append(float(prob_success_um(df)))

        print(file, "is done")
```

```python
final['year'] = year_list
final['semester'] = semester_list
final['lecture_type'] = lecture_type_list
final['prof'] = prof_list
final['name'] = name_list
final['class'] = class_list
final['prob_success'] = prob_success_list
final['prob_success_m'] = prob_success_m_list
final['prob_success_um'] = prob_success_um_list
final['mileage_cut'] = mileage_cut_list
final['mileage_cut_m'] = mileage_cut_m_list
final['mileage_cut_um'] = mileage_cut_um_list
```
