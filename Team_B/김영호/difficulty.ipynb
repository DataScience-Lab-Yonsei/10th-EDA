```python
import os
import pandas as pd
```

```python
def Difficulty(data):
    # data: pandas.DataFrame
    # return: int, 난이도(수강신청 실패 최대 마일리지)
    # --------------------

    data = data[data.iloc[:, -1] == "0"]      # 수강신청 실패

    if len(data) == 0:
        # 수강신청 실패가 없음
        # 1 마일리지만 넣어도 수강신청 성공 -> 수강신청 실패 최대 마일리지 = 1
        max_mileage = 1
    else:
        # 수강신청 실패 최대 마일리지
        max_mileage = max(data[0])

    return int(max_mileage)
```

```python
def Difficulty_byMajorStatus(PATH, names, byMajorStatus=None):
    # PATH: str, Dataset directory path
    # names: list, 전공 수업명
    # byMajorStatus: str, 전공여부
    #                전공(Y(Y)): "4", 복수전공(Y(N)): "3", 비전공(N(N)): "1"
    # return: dict, 전공여부 별 난이도
    #         1(쉬움) ~ 12(어려움)
    # --------------------

    Difficulties = {'2020_2':[], '2021_1':[], '2021_2':[], '2022_1':[], '2022_2':[], '2023_1':[]}

    for name in names:
        for file in os.listdir(os.path.join(PATH, name)):
            DF = pd.read_csv(os.path.join(PATH, name, file), delimiter = ',', header = None)
            DF_byMajorStatus = DF[DF.iloc[:, 2] == byMajorStatus]     # 전공 여부에 따라 DataFrame indexing
            difficulty = Difficulty(DF_byMajorStatus)

            # 해당 학기의 전공여부에 난이도 값 입력
            Difficulties[file[:6]].append(difficulty)

    return Difficulties
```
