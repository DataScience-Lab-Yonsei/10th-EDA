```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def Difficulty(PATH:str, names:str):
    Difficulties = {"Y(Y)":[], "Y(N)":[], "N(N)":[]}

    for name in names:
        for file in os.listdir(os.path.join(PATH, name)):
            DF = pd.read_csv(os.path.join(PATH, name, file), delimiter = ',', header = None)
            DF = DF[DF.iloc[:, -1] == "0"]      # 수강신청 실패
            DF_4 = DF[DF.iloc[:, 2] == "4"]     # 전공자 Y(Y)
            DF_3 = DF[DF.iloc[:, 2] == "3"]     # 복수전공자 Y(N)
            DF_1 = DF[DF.iloc[:, 2] == "1"]     # 비전공자 N(N)

            if len(DF) == 0:
                Difficulties["Y(Y)"].append(1)
                Difficulties["Y(N)"].append(1)
                Difficulties["N(N)"].append(1)
            else:
                if len(DF_4) != 0:
                    Difficulties["Y(Y)"].append(int(float(max(DF_4[0]))))
                    Difficulties["Y(N)"].append(12)
                    Difficulties["N(N)"].append(12)
                elif len(DF_3) != 0:
                    Difficulties["Y(Y)"].append(1)
                    Difficulties["Y(N)"].append(int(float(max(DF_3[0]))))
                    Difficulties["N(N)"].append(12)
                else:
                    Difficulties["Y(Y)"].append(1)
                    Difficulties["Y(N)"].append(1)
                    Difficulties["N(N)"].append(int(float(max(DF_1[0]))))
    
    return Difficulties 
            
```

### 가설검정
각 전공의 난이도 분포가 정규분포를 따른다고 할 수 없고, 등분산이라는 보장이 없으므로, 비모수적인 방법인 Kruskal Wallis 검정을 통해 각 집단의 난이도 분포를 비교한다.(유의수준 0.05)

```python
from scipy import stats
# 전공자 Y(Y) 난이도 비교
# 귀무가설: 세 전공의 전공자의 수강신청 난이도의 분포가 모두 같다.
print(stats.kruskal(BIZ['Y(Y)'], ECO['Y(Y)'], STA['Y(Y)']))

# 귀무가설: BIZ와 ECO의 전공자의 수강신청 난이도의 분포가 같다.
print(stats.kruskal(BIZ['Y(Y)'], ECO['Y(Y)']))

# 귀무가설: BIZ와 STA의 전공자의 수강신청 난이도의 분포가 같다.
print(stats.kruskal(BIZ['Y(Y)'], STA['Y(Y)']))

# 귀무가설: ECO와 STA의 전공자의 수강신청 난이도의 분포가 같다.
print(stats.kruskal(ECO['Y(Y)'], STA['Y(Y)']))
```

```python
# 복수전공자 Y(N) 난이도 비교
# 귀무가설: 세 전공의 복수전공자의 수강신청 난이도의 분포가 모두 같다.
print(stats.kruskal(BIZ['Y(N)'], ECO['Y(N)'], STA['Y(N)']))

# 귀무가설: BIZ와 ECO의 복수전공자의 수강신청 난이도의 분포가 같다.
print(stats.kruskal(BIZ['Y(N)'], ECO['Y(N)']))

# 귀무가설: BIZ와 STA의 복수전공자의 수강신청 난이도의 분포가 같다.
print(stats.kruskal(BIZ['Y(N)'], STA['Y(N)']))

# 귀무가설: ECO와 STA의 복수전공자의 수강신청 난이도의 분포가 같다.
print(stats.kruskal(ECO['Y(N)'], STA['Y(N)']))
```

```python
# 비전공자 N(N) 난이도 비교
# 귀무가설: 세 전공의 비전공자의 수강신청 난이도의 분포가 모두 같다.
print(stats.kruskal(BIZ['N(N)'], ECO['N(N)'], STA['N(N)']))

# 귀무가설: BIZ와 ECO의 비전공자의 수강신청 난이도의 분포가 같다.
print(stats.kruskal(BIZ['N(N)'], ECO['N(N)']))

# 귀무가설: BIZ와 STA의 비전공자의 수강신청 난이도의 분포가 같다.
print(stats.kruskal(BIZ['N(N)'], STA['N(N)']))

# 귀무가설: ECO와 STA의 비전공자의 수강신청 난이도의 분포가 같다.
print(stats.kruskal(ECO['N(N)'], STA['N(N)']))
```
