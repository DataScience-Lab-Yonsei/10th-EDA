```python
drama = pd.read_csv('/content/drive/MyDrive/강건우/통계청 공모전/미니시리즈_메타데이터.csv')
drama = drama[['작품명','장르','방송 채널','방송 기간','방송 시간','방송 횟수','출연자']]
drama
```

```python
drama['방송 시간'].unique()
```

```python
drama['방송 시간'] = drama['방송 시간'].str.replace('매주 월요일, 화요일', '매주 월, 화').str.replace('매주 수요일, 목요일', '매주 수, 목').str.replace('매주 수요일 ~ 목요일', '매주 수, 목').str.replace('매주 월요일', '매주 월').str.replace('매주 월요일 ~ 화요일', '매주 월, 화').str.replace('매주 수요일 목요일', '매주 수, 목').str.replace('매주 화요일', '매주 화').str.replace('매주 수요일','매주 수').str.replace('매주 수요일 ~ 목요일','매주 수, 목').str.replace('화요일 오후 9시 45분','화 오후 9시 45분').str.replace('화요일 오후 10시','화 오후 10시').str.replace('목요일 오후 9시','목 오후 9시').str.replace('목요일 밤 9시', '목 밤 9시').str.replace('화요일 밤 9:00','화 밤 9:00').str.replace('수요일 밤 10:30','수 밤 10:30').str.replace('화요일 밤 9:50','화 밤 9:50').str.replace('월요일, 화요일 밤 22:30','월, 화, 밤 22:30').str.replace('월요일, 화요일 밤 9:50','월, 화, 밤 9:50').str.replace('목요일 오전 12:00','목 오전 12:00').str.replace('금요일 ~ 토요일 오후 9시 50분','금, 토 오후 9시50분').str.replace('월요일, 화요일 밤 10:00 ~ 11:03','월, 화 밤 10:00 ~ 11:03').str.replace('화요일 오후 9시 20분','화 오후 9시 20분').str.replace('TVING: 매주 금요일, 토요일 오전 11:00 선공개 tvN: 매주 수, 목 밤 10:30 ~ 본방송','매주 수, 목 밤 10:30').str.replace('화요일 오후 9시','화 오후 9시').str.replace('일요일 밤 11:05','일 밤 11:05').str.replace('화요일 밤 11:00','화 밤 11:00')
```

```python
drama
```

```python
pattern = r'(월|화|수|목|금|토|일)'
drama['요일'] = drama['방송 시간'].str.findall(pattern).apply(lambda x: ', '.join(x) if isinstance(x, list) else '')
```

```python
drama
```

```python
a = drama['출연자'].dropna()
b = drama.drop(a.index)
b
```

```python
drama = drama.drop(b.index)
drama
```

```python
drama = drama.drop_duplicates()
drama = drama.dropna()
drama
```

```python
import re
pattern = r'(\b\d{1,2}시 \d{1,2}분\b)'
drama['방송 시간'] = drama['방송 시간'].str.replace(pattern, lambda x: x.group().replace('시 ', ':').replace('분', ''))
```

```python
pattern = r'(\b\d{1,2}:\d{1,2}\b)'
drama['시작 시간'] = drama['방송 시간'].str.extract(pattern, expand=False)
```

```python
drama = drama.drop('방송 시간',axis=1)
drama
```

```python
drama['방송 기간'] = drama['방송 기간'].str.replace('시즌2:','').str.replace('2019년 5월 22일 ~ 2019년 7월 11일 MBC 방송 2020년 5월 4일 ~ 2020년 6월 23일 MBC 드라마넷 방송','2019년 5월 22일 ~ 2019년 7월 11일').str.replace('2022년 1월 14일 ~ 2022년 2월 28일 2022년 5월 16일 ~ 2022년 5월 31일','2022년 1월 14일 ~ 2022년 2월 28일')
```

```python
drama['시작일'] = drama['방송 기간'].apply(lambda x: re.findall(r'\d+', x.split('~')[0].strip()))
drama['종료일'] = drama['방송 기간'].apply(lambda x: re.findall(r'\d+', x.split('~')[1].strip()))

drama['시작일'] = drama['시작일'].apply(lambda x: '-'.join(x))
drama['종료일'] = drama['종료일'].apply(lambda x: '-'.join(x))

drama['시작일'] = pd.to_datetime(drama['시작일'], format='%Y-%m-%d')
drama['종료일'] = pd.to_datetime(drama['종료일'], format='%Y-%m-%d')
```

```python
drama = drama.drop('방송 기간',axis=1)
drama
```

```python
drama['요일'] = drama['요일'].str.replace('일,', '')
drama['출연자'] = drama['출연자'].str.replace('외', '').str.replace('外', '')
drama['방송 횟수'] = drama['방송 횟수'].str.replace('부작','')
drama
```

```python
drama['장르'] = drama['장르'].apply(lambda x: x.split(','))
```

```python
dummies_a = pd.get_dummies(drama['장르'].apply(pd.Series).stack()).sum(level=0)
```

```python
drama = drama
drama = drama.join(dummies_a)
drama
```

```python
drama = drama.drop('장르',axis=1)
drama
```

```python
drama['출연자'] = drama['출연자'].apply(lambda x: x.split(','))
```

```python
dummies_b = pd.get_dummies(drama['출연자'].apply(pd.Series).stack()).sum(level=0)
```

```python
drama = drama
drama = drama.join(dummies_b)
drama
```

```python
drama = drama.drop('출연자',axis=1)
```

```python
drama['방송 횟수'] = drama['방송 횟수'].replace('40 + 스페셜 4','40').replace('40 + 스페셜 1','40').replace('24 + 스페셜 2','24').replace('16 + 스페셜 1','16').replace('12 + 스페셜 1','12').replace('시즌2: 14','14').replace('18]]','18')
```

```python
drama['방송 횟수'].unique()
```

```python
drama['요일'] = drama['요일'].apply(lambda x: x.split(','))
```

```python
dummies_c = pd.get_dummies(drama['요일'].apply(pd.Series).stack()).sum(level=0)
```

```python
drama = drama
drama = drama.join(dummies_c)
drama
```

```python
drama = drama.drop('요일',axis=1)
```
