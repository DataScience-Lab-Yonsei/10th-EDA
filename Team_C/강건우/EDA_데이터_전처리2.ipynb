```python
import pandas as pd
import numpy as np
import re
```

전체 데이터 중 작품명/장르/방송채널/방송기간/방송시간/방송횟수/출연자 사용

```python
drama = pd.read_csv('/content/메타데이터.csv')
drama = drama[['작품명','장르','방송 채널','방송 기간','방송 시간','방송 횟수','출연자']]
drama = drama[~drama['방송 채널'].isna()].reset_index(drop = True)
drama = drama.drop_duplicates() # 중복된 행 제거

# 7개 드라마 삭제
drama = drama.dropna().reset_index(drop = True) # 행에서 빈 값이 하나라도 있으면 제거
drama
```

# 작품명 처리

```python
# apply 함수를 사용하여 "_(~~)"" 패턴 제거 적용
drama['작품명'] = drama['작품명'].apply(lambda x: re.sub(r'\ \([^)]*\)', '', x))
# str.replace를 통해 단어 내부에 있는 _를 공백' '으로 변환
drama['작품명'] = drama['작품명'].str.replace('_', ' ')
drama
```

# 방송기간 전처리

```python
# 아래 코드 형식으로 자르기 위해서는 ~ 를 기준으로 문자를 양분해야하는데, 불필요한 단어들이 있음. 이를 삭제.
drama['방송 기간'] = drama['방송 기간'].str.replace('시즌2:','').str.replace('2019년 5월 22일 ~ 2019년 7월 11일 MBC 방송 2020년 5월 4일 ~ 2020년 6월 23일 MBC 드라마넷 방송','2019년 5월 22일 ~ 2019년 7월 11일').str.replace('2022년 1월 14일 ~ 2022년 2월 28일 2022년 5월 16일 ~ 2022년 5월 31일','2022년 1월 14일 ~ 2022년 2월 28일').str.replace('every1', '')
```

```python
drama.loc[drama['방송 기간'].str.split('~').str.len() < 2, '방송 기간'] = ['2022년 6월 3일 ~ 2022년 7월 23일', '2023년 2월 7일 ~ 2022년 3월 14일']
drama.iloc[238, 3] = '2021년 9월 17일 ~ 2021년 10월 30일'
drama.iloc[18, 3] = '2019년 9월 20일 ~ 2019년 11월 23일'
drama.iloc[[42, 44, 85], 3] = ['2019년 2월 15일 ~ 2019년 4월 20일', '2019년 4월 26일 ~ 2019년 7월 13일', '2019년 6월 1일 ~ 2019년 9월 7일']
```

```python
drama['시작일'] = drama['방송 기간'].apply(lambda x: re.findall(r'\d+', x.split('~')[0].strip()))
drama['종료일'] = drama['방송 기간'].apply(lambda x: re.findall(r'\d+', x.split('~')[1].strip()))

drama['시작일'] = drama['시작일'].apply(lambda x: '-'.join(x))
drama['종료일'] = drama['종료일'].apply(lambda x: '-'.join(x))

drama['시작일'] = pd.to_datetime(drama['시작일'], format='%Y-%m-%d')

mod_idx = drama.loc[(drama['종료일'].str.len() != 10) & (drama['종료일'].str.len() != 9) & (drama['종료일'].str.len() != 8)].index.tolist()
mods =  ['2019-9-7', '2020-10-24', '2021-1-12', '2020-6-12', '2020-10-13', '2020-8-8', '2021-3-23', '2021-7-4', '2021-8-13','2022-6-12',
         '2022-9-10', '2023-10-16', '2023-08-06', '2023-7-29', '2023-8-13', '2020-4-11', '2023-2-24']
drama.iloc[mod_idx, -1] = mods

drama['종료일'] = pd.to_datetime(drama['종료일'], format='%Y-%m-%d')

# 전처리에 사용한 방송 기간 열 삭제
drama = drama.drop('방송 기간',axis=1)
```

```python
# 시작일 기준으로 정렬
drama = drama.sort_values(by = '시작일').reset_index(drop = True)
drama
```

```python
pattern = r'(월|화|수|목|금|토|일)'
drama['요일'] = drama['방송 시간'].str.findall(pattern).apply(lambda x: ''.join(x) if isinstance(x, list) else '')
drama.loc[drama['방송 시간'].str.contains('평일'), '요일'] = '평일'
```

```python
drama.loc[drama['요일'] == '', '요일'] = ['수목', '토일', '토일', '토일', '금토']
```

```python
# drama['요일'].unique()
stranges = ['수일', '목일', '수목월화', '금토월일월일', '수목수', '월화월화', '화월', '수목수목', '월화일', '화화','일월일', '금토금토', '토일토일', '금월화', '월일월일', '월금', '금토수목', '금토일']
mods = ['수일', '목일', '수목', '금토', '수목', '월화', '화', '수목', '월화', '화', '화', '일', '금토', '토일', '월화', '금토', '월금', '수목', '금토일']
drama.loc[drama['요일'].isin(stranges), '요일'] = mods
```

```python
drama.loc[drama['요일'] == '평일', '요일'] = '월화수목금'
```

## 요일 원핫 인코딩

```python
dummy_df = drama['요일'].str.get_dummies(sep='')
dummy_df = dummy_df[['월', '화', '수', '목', '금', '토', '일']]
drama = pd.concat([drama, dummy_df], axis = 1)
drama
```

## 방송 시작 시간

```python
# 아침 드라마 제거
drama = drama.loc[~(drama['방송 시간'].str.contains('아침')) | (drama['방송 시간'].str.contains('본방송'))]
```

```python
# 방송 시간 적기
pattern = r'(\b\d{1,2}시 \d{1,2}분\b)'
drama['방송 시간'] = drama['방송 시간'].str.replace(pattern, lambda x: x.group().replace('시 ', ':').replace('분', ''))

pattern = r'(\b\d{1,2}시\b)'
drama['방송 시간'] = drama['방송 시간'].str.replace(pattern, lambda x: re.sub(r'\d{1,2}', lambda y: y.group() + ':00 ', x.group()))
drama
```

```python
# 방송 시작 시간
pattern = r'(\b\d{1,2}:\d{1,2}\b)'
drama['시작 시간'] = drama['방송 시간'].str.extract(pattern, expand=False)

drama = drama.drop('방송 시간',axis=1)
drama
```

# 장르 원핫 인코딩

```python
drama['장르'] = drama['장르'].apply(lambda x: x.split(','))
dummies_a = pd.get_dummies(drama['장르'].apply(pd.Series).stack()).sum(level=0)

drama = drama
drama = drama.join(dummies_a)
drama
```

# 출연자 처리

```python

drama['출연자'] = drama['출연자'].str.replace('외', '').str.replace('外', '').str.replace('등', '')
drama['출연자'] = drama['출연자'].apply(lambda x: x.split(','))
dummies_b = pd.get_dummies(drama['출연자'].apply(pd.Series).stack()).sum(level=0)
drama = drama
drama = drama.join(dummies_b)
drama
```

# 방송 횟수 처리

```python
# 스페셜이라 써있는 것도 있고 스폐셜이라고 쓰여있는 것도 있어서 2가지 형식을 or로 묶음
drama['방송 횟수'] = drama['방송 횟수'].str.replace((r'(스페셜 \d부작)|(스폐셜 \d부작)'), '', regex=True)

# 부작 앞의 2글자가 방송횟수와 같음
drama['방송 횟수'] = drama['방송 횟수'].apply(lambda x: x.split('부작')[0][-2:] if '부작' in x else '')
drama
```

```python
drama[drama['방송 횟수'] == ''] = 12
drama['방송 횟수'] = drama['방송 횟수'].astype(int)
```

# 데이터 출력
