```python
import pandas as pd
import numpy as np
import re
```

전체 데이터 중 작품명/장르/방송채널/방송기간/방송시간/방송횟수/출연자 사용

```python
drama = pd.read_csv('/content/미니시리즈_메타데이터.csv')
drama = drama[['작품명','장르','방송 채널','방송 기간','방송 시간','방송 횟수','출연자']]
drama = drama[~drama['방송 채널'].isna()].reset_index(drop = True)
drama
```

```python
drama = drama.drop_duplicates() # 중복된 행 제거

# 대략 18개 행 삭제
drama = drama.dropna() # 행에서 빈 값이 하나라도 있으면 제거
drama
```

# 작품명 처리

```python
# apply 함수를 사용하여 "_(~~)"" 패턴 제거 적용
drama['작품명'] = drama['작품명'].apply(lambda x: re.sub(r'\_\([^)]*\)', '', x))
# str.replace를 통해 단어 내부에 있는 _를 공백' '으로 변환
drama['작품명'] = drama['작품명'].str.replace('_', ' ')
```

```python
drama
```

# 방송기간 전처리

```python
# 아래 코드 형식으로 자르기 위해서는 ~ 를 기준으로 문자를 양분해야하는데, 불필요한 단어들이 있음. 이를 삭제.
drama['방송 기간'] = drama['방송 기간'].str.replace('시즌2:','').str.replace('2019년 5월 22일 ~ 2019년 7월 11일 MBC 방송 2020년 5월 4일 ~ 2020년 6월 23일 MBC 드라마넷 방송','2019년 5월 22일 ~ 2019년 7월 11일').str.replace('2022년 1월 14일 ~ 2022년 2월 28일 2022년 5월 16일 ~ 2022년 5월 31일','2022년 1월 14일 ~ 2022년 2월 28일')
```

```python
drama['시작일'] = drama['방송 기간'].apply(lambda x: re.findall(r'\d+', x.split('~')[0].strip()))
drama['종료일'] = drama['방송 기간'].apply(lambda x: re.findall(r'\d+', x.split('~')[1].strip()))

drama['시작일'] = drama['시작일'].apply(lambda x: '-'.join(x))
drama['종료일'] = drama['종료일'].apply(lambda x: '-'.join(x))

drama['시작일'] = pd.to_datetime(drama['시작일'], format='%Y-%m-%d')
drama['종료일'] = pd.to_datetime(drama['종료일'], format='%Y-%m-%d')

# 전처리에 사용한 방송 기간 열 삭제
drama = drama.drop('방송 기간',axis=1)
```

```python
# 시작일 기준으로 정렬
drama = drama.sort_values(by = '시작일').reset_index(drop = True)
drama
```

```python
pattern = r'(월|화|수|목|금|토|일)'
drama['요일'] = drama['방송 시간'].str.findall(pattern).apply(lambda x: ', '.join(x) if isinstance(x, list) else '')
```

```python
# 노가다 요소
drama.loc[drama['작품명'] == '머니게임', '요일'] = '수, 목' # 머니게임 요일 넣기
```

## 요일 원핫 인코딩

```python
# 두번째 콤마 앞의 부분만 추출, 콤마가 1개 이하면 단어 전부 반환

def extract_day(text):
    commas = text.count(',')
    if commas >= 2:
        second_comma_index = text.find(',', text.find(',') + 1)
        return text[:second_comma_index].strip()
    else:
        return text.strip()

drama['요일'] = drama['요일'].apply(extract_day)
```

```python
# dummies_c = pd.get_dummies(drama['요일'].apply(pd.Series).stack()).sum(level=0)
dummies_c = pd.get_dummies(drama['요일'])
# drama = drama
drama = drama.join(dummies_c)
drama
```

## 방송 시작 시간

```python
# 방송 시간 적기
pattern = r'(\b\d{1,2}시 \d{1,2}분\b)'
drama['방송 시간'] = drama['방송 시간'].str.replace(pattern, lambda x: x.group().replace('시 ', ':').replace('분', ''))

pattern = r'(\b\d{1,2}시\b)'
drama['방송 시간'] = drama['방송 시간'].str.replace(pattern, lambda x: re.sub(r'\d{1,2}', lambda y: y.group() + ':00 ', x.group()))
drama
```

```python
# 방송 시작 시간
pattern = r'(\b\d{1,2}:\d{1,2}\b)'
drama['시작 시간'] = drama['방송 시간'].str.extract(pattern, expand=False)

drama = drama.drop('방송 시간',axis=1)
drama
```

# 출연자 처리

```python

drama['출연자'] = drama['출연자'].str.replace('외', '').str.replace('外', '')
drama['출연자'] = drama['출연자'].apply(lambda x: x.split(','))
```

```python
dummies_b = pd.get_dummies(drama['출연자'].apply(pd.Series).stack()).sum(level=0)
```

```python
drama = drama
drama = drama.join(dummies_b)
drama
```

# 방송 횟수 처리

```python
# 스페셜이라 써있는 것도 있고 스폐셜이라고 쓰여있는 것도 있어서 2가지 형식을 or로 묶음
drama['방송 횟수'] = drama['방송 횟수'].str.replace((r'(스페셜 \d부작)|(스폐셜 \d부작)'), '', regex=True)

# 부작 앞의 2글자가 방송횟수와 같음
drama['방송 횟수'] = drama['방송 횟수'].apply(lambda x: x.split('부작')[0][-2:] if '부작' in x else '')
drama
```

```python
drama['방송 횟수'] = drama['방송 횟수'].astype(float)
```
