```python
# 수정본

# SF, 가족, 로맨스, 미스터리, 스릴러(범죄, 느와르 관련)
# 액션, 역사, 추리, 코미디, 판타지, 휴먼

genre_groups = {
    '액션': ['액션', '스포츠', '전쟁'],
    '코미디': ['코미디', '로맨틱코미디', '로코', '로맨틱 코미디', '로맨틱 코미디 드라마',
                '코믹', '블랙 코미디', '코믹 미스터리 사극 드라마'],
    '드라마': ['드라마', '멜로', '로맨스', '로맨스 드라마', '로맨스드라마', '로맨틱드라마',
                '멜로 로맨스 코미디', '멜로멜로', '멜로 로맨스'],
    '스릴러': ['스릴러', '범죄', '법률', '범죄물', '법정', '블록버스터', '블랙코미디',
                '사이언스 액션 히어로', '서스펜스', '서스펜스추리', '피카레스크'],
    '공포': ['공포', '호러', '금융'],
    '판타지': ['판타지', '시대극', '판타지 드라마', '판타지로코 드라마'],
    '가족': ['가족', '어드벤처', '첩보', '뮤지컬'],
    '미스터리': ['미스터리', '미스터리 멜로', '미스터리 스릴러', '미스터리 수사물'],
    '전쟁': ['전쟁', '사극'],
    'SF': ['SF', '엑소시즘', '타임 크로싱', '프라이빗 오피스 멜로', '하드보일드', '정치',
                '정치 스릴러', '전쟁 스릴러', '슬래셔 스릴러']
}
```

```python
# genre_groups 딕셔너리에 정의된 각 그룹에 속하는 열들을 찾아서 grouped_genres에 저장
grouped_genres = {}
for group, genres in genre_groups.items():
    grouped_genres[group] = [genre for genre in meta.columns if any(g in genre for g in genres)]
```

```python
#grouped_genres 딕셔너리에 저장된 장르 그룹들과
#그에 해당하는 열들을 사용하여 새로운 데이터프레임 만들
grouped_df = pd.DataFrame()
for group, genres in grouped_genres.items():
    grouped_df[group] = meta[genres].sum(axis=1)
```

```python
grouped_df
```

```python
#작품명과 기타 데이터 살리기
sliced_meta = meta.iloc[:, :5]
sliced_meta
```

```python
merged_df = pd.concat([sliced_meta, grouped_df], axis=1)
merged_df
```

```python
#각 셀의 원핫인코딩을 살리기 위해 2이상인 것들은 1로 대체함
def replace(x):
    try:
        numeric = pd.to_numeric(x)
    except:
        numeric = float('nan')

    if pd.notnull(numeric) and numeric >= 2:
        return 1
    else:
        return x
```

```python
merged_df = merged_df.applymap(replace)
merged_df
```
