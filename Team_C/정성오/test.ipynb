## preprocessing
결측치 제거 및 rating_d의 방송일자 column을 datetime 형식으로 바꿈

```python
# 결측치 제거
rating_d = rating_d.dropna(axis = 0).reset_index(drop=True)
rating_d.shape
```

```python
# '방송일자'열의 dtype이 string으로 나오기 때문에, datetime으로 변환시키기
rating_d['방송일자'] = pd.to_datetime(rating_d['방송일자'])
print(rating_d.dtypes)
```

```python
# 12부작 ~ 16부작인 드라마만 추출

def drama_episode_extract(rating_d, min_episode, max_episode):
    #드라마별 회차 추출
    drama_episode_count = rating_d.groupby('작품명')['시청률'].count().reset_index()
    drama_episode_count.rename(columns={'시청률':'count'}, inplace = True)
    drama_episode_count.head()

    # min_episode이상, max_episode 미만의 드라마만 추출
    bool_mask = (min_episode<=drama_episode_count['count']) & (drama_episode_count['count']<=max_episode)
    extracted_dama_df = drama_episode_count[bool_mask].reset_index(drop=True)
    
    extracted_dama_ratig = rating_d[rating_d['작품명'].isin(extracted_dama_df['작품명'])].reset_index(drop=True)
    extracted_dama_ratig
    
    return extracted_dama_ratig


min_episode = 12
max_episode = 16
extracted_dama_ratig = drama_episode_extract(rating_d, min_episode, max_episode)
extracted_dama_ratig
```

## 시청률 top 10 드라마 추출

```python
# 드라마별 평균 시청률 계산
mean_rating_df = extracted_dama_ratig.groupby("작품명")['시청률'].mean().reset_index()
# 시청률 높은 순서대로 정렬
mean_rating_df = mean_rating_df.sort_values(by='시청률', ascending = False)
mean_rating_df
```

```python
# 시청률 top10 드라마 추출

top_number = 10 # 상위 몇 개까지 추출할지 결정

top_n_drama_df = mean_rating_df.iloc[:top_number,:].reset_index(drop=True)
top_n_drama_df
```

```python
# 드라마별 방송 기간 추출
def attach_drama_broadcasting_period(top_n_drama_df, drama_list_d):
    top_n_drama_df['시작'] = datetime.date(year=2000, month=1, day=1)
    top_n_drama_df['종영'] = datetime.date(year=2000, month=1, day=1)

    for i, title in enumerate(top_n_drama_df['작품명']):
        drama = drama_list_d[drama_list_d['작품명'] == title]
        begin, end = drama['방송 기간'].values[0].split(',')
        top_n_drama_df.loc[i,'시작'] =\
                        datetime.datetime.strptime(begin.strip(), '%Y.%m.%d').date()
        top_n_drama_df.loc[i,'종영'] =\
                        datetime.datetime.strptime(end.strip(), '%Y.%m.%d').date()
    return top_n_drama_df
    
    
top_n_drama_df = attach_drama_broadcasting_period(top_n_drama_df, drama_list_d)
top_n_drama_df
```

```python
# 시작일과 종영일이 바뀐 경우, 원래대로 바꿔주기
def swap_wrong_date(top_n_drama_df: pd.DataFrame)->pd.DataFrame:
    # 시작일이 종영일보다 뒤인 경우(즉, 둘이 바뀌어서 잘못 추출된 경우) 추출
    bool_mask = top_n_drama_df['시작'] > top_n_drama_df['종영']
    # 원래 '시작', '종영' 순서로 열이 되어 있지만 추후에 둘을 바꾸기 위해 순서 바꿔서 추출
    end_start_date =  top_n_drama_df[bool_mask][['종영', '시작']]
    top_n_drama_df.loc[bool_mask, ['시작','종영']] = end_start_date.values
    return top_n_drama_df


top_n_drama_df = swap_wrong_date(top_n_drama_df)
top_n_drama_df
```

## 드라마별 출연자 데이터 만들기

```python
# top_n_drama_df에 '출연자' 열을 추가
def add_actor_column(top_n_drama_df:pd.DataFrame, meta_df:pd.DataFrame)->pd.DataFrame:
    actor_col = []
    for title in top_n_drama_df['작품명']:
        meta_df = meta_d[['작품명','출연자']]
        actors = meta_df[meta_df['작품명'] == title]['출연자'].values[0]
        actor_col.append(actors)

    top_n_drama_df['출연자'] = pd.Series(actor_col)
    return top_n_drama_df


top_n_drama_df = add_actor_column(top_n_drama_df, meta_d)
top_n_drama_df
```
