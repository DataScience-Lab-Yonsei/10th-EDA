```python
#merged csv 불러오기
path = "/content/drive/MyDrive/eda/merged.csv"
cols = ["가입자 일련번호", "성별코드", "연령대코드", "시도코드", "주상병코드", '요양일수', '입내원일수', '심결가산율', '심결요양급여비용총액', '심결본인부담금',
       '심결보험자부담금', '총처방일수']
df_raw = pd.read_csv(path, usecols = cols, encoding="utf-8")

#데이터프레임 복사
temp = df_raw.copy(deep = True)

#메모리 free
del df_raw
gc.collect()

#["주상병코드"] 데이터 대분류로 축소
temp["주상병코드"] = temp["주상병코드"].str.slice(0,1)

#시는 si, 도는 do로 변환
temp["시도코드"] = temp["시도코드"].replace(
    {11 : "si"
    ,26 : "si"
    ,27 : "si"
    ,28 : "si"
    ,29 : "si"
    ,30 : "si"
    ,31 : "si"
    ,36 : "si"
    ,41 : "do"
    ,42 : "do"
    ,43 : "do"
    ,44 : "do"
    ,45 : "do"
    ,46 : "do"
    ,47 : "do"
    ,48 : "do"
    ,49 : "do"})

#주상병코드, 시도코드 원핫인코딩
temp = pd.get_dummies(temp, columns=["주상병코드", "시도코드"])

#Normalization 해줘야하는 칼럼 리스트로 정리
norm_cols = ["성별코드", "연령대코드", "요양일수", "입내원일수", "심결가산율", "심결요양급여비용총액", "심결본인부담금", "심결보험자부담금", "총처방일수"]

#normalization 수행
for column in norm_cols:
  data = temp[column]
  temp[column] = minmax_scale(data)

#normalization 결과 확인
temp
```

## 메모리 초과 방지를 위해 샘플추출

```python
#재현을 위한 random_state 지정 및 샘플 1,000,000개 추출
sample = temp.sample(n=1000000, random_state=777)
```

## PCA 수행

```python
#Scree plot으로 파악한 5개의 PC 산출
pca = PCA(n_components=5)
principle_componenets = pca.fit_transform(sample)

#새로운 DataFrame 선언
df_pc = pd.DataFrame(data = principle_componenets, columns=['PC1', 'PC2','PC3','PC4','PC5'])

#새 DataFrame 확인
df_pc
```

```python
#메모리 free
del sample
gc.collect()
```

## K-means clustering 시행

```python
# 실험할 범위 지정
k_range = list(range(2, 30+1))

# 각 metric을 담을 list 선언
calinski_harabasz_scores = []
davies_bouldin_scores = []
sse = []

# 2개~20개 까지의
for k in tqdm(k_range):
    # Create an instance of KMeans with the current value of n_clusters
    model = KMeans(n_clusters=k)

    # Fit the model to the data
    model.fit(df_pc)

    # Calculate the Calinski-Harabasz score for the current value of k
    calinski_harabasz_scores.append(calinski_harabasz_score(df_pc, model.labels_))

    # Calculate the Davies-Bouldin score for the current value of k
    davies_bouldin_scores.append(davies_bouldin_score(df_pc, model.labels_))

    # Calculate the SSE for the current values of k
    sse.append(model.inertia_)
```

```python
# Plot the results
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(50, 15))

ax1.plot(k_range, calinski_harabasz_scores)
ax1.set_xlabel('Number of Clusters')
ax1.set_ylabel('Calinski-Harabasz Score')
ax1.set_title('Calinski-Harabasz Method')

ax2.plot(k_range, davies_bouldin_scores)
ax2.set_xlabel('Number of Clusters')
ax2.set_ylabel('Davies-Bouldin Score')
ax2.set_title('Davies-Bouldin Method')

ax3.plot(k_range, sse, 'bx-')
ax3.set_xlabel('k')
ax3.set_ylabel('Sum of squared distances')
ax3.set_title('Elbow Method For Optimal k')



plt.show()
```
