## 필요 라이브러리 및 데이터셋 로드

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
```

```python
#클러스터링 결과 로드
path = "/Users/younyung.gene/__DSL__/EDA/_pc5_16clusters_/KMeans_16.csv"
df = pd.read_csv(path, encoding = "utf-8")
```

```python
#데이터 확인
df.info()
```

## 시각화

```python
#feature 정리
pc = ["PC1", "PC2", "PC3", "PC4","PC5"]

# 10개의 scatterplot 작도
for idx, componenet in (enumerate(pc[:-1])):
    x_data = df[componenet]
    remaining = pc[idx+1:]
    for versus in remaining:
        y_data = df[versus]
        plt.scatter(x_data, y_data, c = df["cluster"], cmap="viridis")
        plt.title(f"{componenet} vs {versus} 16 cluster visualization")
        plt.show()
```
