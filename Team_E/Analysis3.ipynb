```python
import pandas as pd
import folium
from folium import Marker
from folium.plugins import MarkerCluster
```

```python
df = pd.read_csv('/content/지하철역_GEOM (역사마스터).csv', encoding = 'cp949')
df.head()
```

```python
map = folium.Map(location=[37.513011, 127.053282], tiles="OpenStreetMap", zoom_start=11)
map
```

```python
for _, i in df.iterrows():
    Marker(location = [i['환승역X좌표'], i['환승역Y좌표']],
           popup = "호선명칭: " + str(i['호선명칭']) + "<br>" + "역한글명칭: " + str(i['역한글명칭']) + "<br>",
           icon = folium.Icon(color = 'blue')
           ).add_to(map)
```

```python
for _, i in df.iterrows():
    Marker(location = [i['환승역X좌표'], i['환승역Y좌표']],
           popup = "<pre>" + "호선명칭: " + str(i['호선명칭']) + "<br>" + "역한글명칭: " + str(i['역한글명칭']) + "<br>" + "</pre>",
           icon = folium.Icon(color = 'blue')
           ).add_to(map)
```

```python
map
```

```python
cluster = MarkerCluster()
for _, i in df.iterrows():
    cluster.add_child(
        Marker(location = [i['환승역X좌표'], i['환승역Y좌표']],
               popup = "<pre>" + "호선명칭: " + str(i['호선명칭']) + "<br>" + "역한글명칭: " + str(i['역한글명칭']) + "<br>" + "</pre>",)
    ).add_to(map)
```
